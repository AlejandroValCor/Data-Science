```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Fill in the line below: Specify the path of the CSV file to read
my_filepath = "./Datasets/daily_food_nutrition_dataset.csv"

# Fill in the line below: Read the file into a variable my_data
my_data = pd.read_csv(my_filepath, index_col='User_ID', parse_dates=True)
```

```python
# Print the first five rows of the data
my_data.head()
```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=my_data['Water_Intake (ml)'])
```

```python
plt.figure(figsize=(16,6))

sns.boxplot(data=my_data, x="Meal_Type", y="Calories (kcal)")
```

```python
plt.figure(figsize=(16,6))

sns.boxplot(data=my_data, x="Category", y="Protein (g)")
sns.boxplot(data=my_data, x="Category", y="Fat (g)")
sns.boxplot(data=my_data, x="Category", y="Carbohydrates (g)")
```

```python
plt.figure(figsize=(16,6))

my_data['Date'] = pd.to_datetime(my_data['Date'])
daily_calories = my_data.groupby('Date')["Calories (kcal)"].sum().reset_index()
sns.lineplot(data=daily_calories, x="Date", y="Calories (kcal)")
```

```python
plt.figure(figsize=(16,6))

sns.scatterplot(data=my_data, x="Fat (g)", y="Calories (kcal)", hue="Category")
```
