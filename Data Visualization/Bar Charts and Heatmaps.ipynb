```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Path of the file to read
ign_filepath = "./Datasets/ign_scores.csv"

# Fill in the line below to read the file into a variable ign_data
ign_data = pd.read_csv(ign_filepath, index_col='Platform')
```

```python
ign_data
```

```python
# Fill in the line below: What is the highest average score received by PC games,
# for any genre?
high_score = ign_data.loc['PC'].max()

# Fill in the line below: On the Playstation Vita platform, which genre has the 
# lowest average score? Please provide the name of the column, and put your answer 
# in single quotes (e.g., 'Action', 'Adventure', 'Fighting', etc.)
worst_genre = ign_data.loc['PlayStation Vita'].idxmin()

print(high_score)
print(worst_genre)
```

```python
# Bar chart showing average score for racing games by platform
plt.figure(figsize=(16,10))

plt.title("Average score for racing games by console")

sns.barplot(x=ign_data['Racing'], y=ign_data.index, palette="Set2")

plt.xlabel("Score")
```

```python
highestScorePlatform = ign_data['Racing'].idxmax()
lowestScorePlatform = ign_data['Racing'].idxmin()

print("Platform with highest average score in racing games: ", highestScorePlatform)
print("Platform with lowest average score in racing games: ", lowestScorePlatform)
```

```python
# Heatmap showing average game score by platform and genre
plt.figure(figsize=(10,10))

plt.title("Heatmap of average game scores by genre and platform")

sns.heatmap(ign_data, annot=True)

plt.xlabel("Genre")
```

```python
lowestScoreGenres = list(ign_data.columns[ign_data.min() == ign_data.min().min()])
lowestScorePlatform = ign_data.stack().idxmin()[0]

print("Platform: ", lowestScorePlatform)
print("Has the lowest score on genres: ", lowestScoreGenres)

highestScoreGenres = list(ign_data.columns[ign_data.max() == ign_data.max().max()])
highestScorePlatform = ign_data.stack().idxmax()[0]

print("Platform: ", highestScorePlatform)
print("Has the highest score on genres: ", highestScoreGenres)
```
